```python
import task_4
import pandas as pd
import time
from collections.abc import Callable
import tracemalloc
```

```python
MAX_TIME = 5
```

```python
def experiment(fib: Callable[[int], int], duration: int = MAX_TIME):
    tracemalloc.start()
    
    start_time = time.time()
    n = 2
    result = None

    while (time.time() - start_time) < duration:
        try:
            result = fib(n)
            n += 1
        except RecursionError:
            print(f"RecursionError для {fib.__name__}")
            result = "Ошибка: превышена глубина рекурсии"
            break

    end_time = time.time()
    final_time = end_time - start_time
    
    _, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    
    peak = f"{peak / (2**11)} МБ"

    return result, final_time, peak
```

```python
df = pd.DataFrame(
    {
        "Рекурсия без кэш": experiment(task_4.fib_basic),
        "Рекурсия с кэш": experiment(task_4.fib_cash),
        "Итеративно со списком": experiment(task_4.fib_iter_list),
        "Итеративно на лету": experiment(task_4.fib_iter_fly),
    },
    index=["Значение", f"Затраченное время при ограничении {MAX_TIME} сек", "Пиковый расход памяти"],
    dtype=object
)
```
